# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
ReLU = lambda x: np.maximum(0, x)

x = np.arange(-10, 10, 0.1)

plt.plot(x, ReLU(x), 'b')
plt.grid()
plt.title('Rectified Linear Unit - ReLU')
plt.show()
```

```python
dReLU = lambda x: np.heaviside(x, 1)

plt.plot(x, dReLU(x), 'r')
plt.grid()
plt.title('dReLU')
plt.show()
```
